```python
df = spark.read.csv("/Volumes/factory_smartification/default/smart_manufecturing", header=True, inferSchema=True)
df.show()
print("PySpark is working")
```

 Check Data Quality

```python
from pyspark.sql.functions import col, sum as _sum

df.select([
    _sum(col(c).isNull().cast("int")).alias(c)
    for c in df.columns
]).show()


df.describe().show()

df.groupBy("anomaly_flag").count().show()

df.write.mode("overwrite").saveAsTable("machine_cleaned_data")
```
